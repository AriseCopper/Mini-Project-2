```python
import tkinter as tk
import random
WIDTH = 400
HEIGHT = 400
SNAKE_SIZE = 20
snake_speed = 100 
snake = [(100, 100), (80, 100), (60, 100)]
snake_direction = "Right"

food_pos = (200, 200)

def change_direction(new_direction):
    global snake_direction
    opposites = {"Right": "Left", "Left": "Right", "Up": "Down", "Down": "Up"}
    if snake_direction != opposites.get(new_direction):
        snake_direction = new_direction

def move_snake():
    global food_pos

    head_x, head_y = snake[0]

    if snake_direction == "Right":
        head_x += SNAKE_SIZE
    elif snake_direction == "Left":
        head_x -= SNAKE_SIZE
    elif snake_direction == "Up":
        head_y -= SNAKE_SIZE
    elif snake_direction == "Down":
        head_y += SNAKE_SIZE

    new_head = (head_x, head_y)

    if (head_x < 0 or head_x >= WIDTH or head_y < 0 or head_y >= HEIGHT or new_head in snake):
        canvas.create_text(WIDTH/2, HEIGHT/2, text="Game Over!", fill="red", font=("Arial", 20, "bold"))
        return

    snake.insert(0, new_head)

    if new_head == food_pos:
        food_pos = (random.randint(0, (WIDTH-SNAKE_SIZE)//SNAKE_SIZE)*SNAKE_SIZE,
                    random.randint(0, (HEIGHT-SNAKE_SIZE)//SNAKE_SIZE)*SNAKE_SIZE)
    else:
        snake.pop()

    draw_snake()
    root.after(snake_speed, move_snake)

def draw_snake():
    canvas.delete("all")
    canvas.create_rectangle(food_pos[0], food_pos[1], food_pos[0]+SNAKE_SIZE, food_pos[1]+SNAKE_SIZE, fill="red")
    for x, y in snake:
        canvas.create_rectangle(x, y, x+SNAKE_SIZE, y+SNAKE_SIZE, fill="lime")

root = tk.Tk()
root.title("üêç Snake Game")
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="black")
canvas.pack()

root.bind("<Right>", lambda e: change_direction("Right"))
root.bind("<Left>", lambda e: change_direction("Left"))
root.bind("<Up>", lambda e: change_direction("Up"))
root.bind("<Down>", lambda e: change_direction("Down"))

draw_snake()
move_snake()

root.mainloop()
```
